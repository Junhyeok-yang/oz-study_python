```python
#3. while문 조건식 증감, 감소

#1. 기본적인 증가 유형

i = 1
while i < 10:
    print('이정도로 성장할 줄이야', i)
    i += 1

#2. 기본적인 감소 유형
i = 10
while i > 0:
    print('이정도로 성장할 줄이야', i)
    i -= 1
```

```python
#6.while문에도 else가 있다

양파 = 10
while 양파 > 10:
    print('양파', 양파, '개 남음')
    if 양파 == 0:
        break
    양파 -= 1
else:
    print('양파 다 팔림')
```

```python
#7.반복문에 흐름을 제어하는 continue와 break

i = 0
while True:
    print(i)
    i += 1
    if i == 10:
        break

for i in range(100):
    print(i)
    if i == 10:
        break
```

```python
i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue
    print(i)

for i in range(10):
     if i % 2 == 0:
        continue
     print(i)
```
