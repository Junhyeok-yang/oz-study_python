###3. 함수의 반환
함수는 **return을 사용하여 값을 반환**합니다.  
함수 내부의 **return을 만나면, 이후의 코드를 실행하지 않고 함수를 종료**합니다.

````python
#함수를 선언합니다.
def function(x, y):
	z = x + y
	#함수의 값을 return합니다.
	return z

#옥수수 10개, 수박 2개
#function(10, 2) == z
print('옥수수와 수박의 수확량 :', function(10, 2))

#출력
# 옥수수와 수박의 수확량 : 12
````

###5. return 유무의 차이
**return이 없어도 함수를 실행할 수 있습니다.**

다만 **return이 없는 경우에는, 함수 외부에서 함수 내부의 값을 얻을 수 없습니다.**

````python
def function(x, y):
	z = x + y
	print(z)

#return이 없이 때문에 z의 값을 함수 외부에서 얻을 수 없습니다.

print('옥수수와 수박의 수확량 :', function(10, 2))

#출력
# 12
# 옥수수와 수박의 수확량 : None
````

````python
def 계산(가격, 개수):
	return 가격 * 개수

#return이 있을 경우에는 함수 외부에서 값을 사용할 수 있습니다.

print(계산(1000, 5), '원입니다.')

#출력
#5000원입니다.
````

###6. 함수 내에서 전역변수 사용하기
**함수 밖에 있는 전역 변수의 값을 함수 내부에서 변경하려면 global 키워드**를 사용해 선언해야합니다.

````python
# 전역 변수 a를 100으로 초기화합니다.
a = 100

# 함수를 정의합니다.
def f():
    # 함수 내에서 전역 변수 a를 사용하기 위해 global 키워드를 사용합니다.
    global a
    # 전역 변수 a의 값을 1 증가시킵니다.
    a = a + 1
    # 증가된 전역 변수 a의 값을 반환합니다.
    return a

# 함수 f를 호출하여 전역 변수 a의 값을 1 증가시킵니다.
f()

# 증가된 값을 출력합니다.
print(a)

# 출력:
#101
````

**dict타입에 한해서 전역변수에 있는 값을 지역변수로 가져와 사용**할 수 있다.

````python
# 등급별 가격을 dict로 정의합니다.
won = {"A등급": 1000, "B등급": 500, "C등급": 100}

# 등급별 가격을 합산하여 반환하는 함수를 정의합니다.
def 계산(a, b, c):
    # 딕셔너리의 값들을 합하여 등급별 가격의 합계를 계산합니다.
    합계 = won['A등급'] * a + won['B등급'] * b + won['C등급'] * c
    return 합계

# 계산 함수를 호출하여 등급별 가격의 합계를 출력합니다.
print(계산(5, 2, 3), '원입니다.')

#출력
#6300원입니다.
````

**전역 변수를 아규먼트로 사용하여 함수 내부에 값을 전달**할 수 있습니다.

````python
# 전역 변수로 원주율을 정의합니다.
PI = 3.14

# 원의 넓이를 계산하는 함수를 정의합니다. 함수는 원의 반지름과 원주율을 매개변수로 받습니다.
def circle(r, inputpi):
    # 원의 넓이를 계산하여 변수 z에 저장합니다.
    z = r * r * inputpi
    # 계산된 넓이 값을 반환합니다.
    return z

# circle 함수를 호출하여 반지름이 10인 원의 넓이를 계산합니다. 이때 사용할 원주율을 전역 변수 PI로 지정합니다.
result = circle(10, PI)

# 계산된 넓이를 출력합니다.
print(result)
````

더 복잡한 함수

````python
# 각각의 수확물을 저장할 전역 변수를 초기화합니다.
A = 0  # 감자
B = 0  # 양파
C = 0  # 옥수수
유물 = 0  # 기타 수확물

# 감자를 수확하는 함수를 정의합니다.
def 감자():
    # 전역 변수 A에 1을 더합니다.
    global A
    A += 1

# 양파를 수확하는 함수를 정의합니다.
def 양파():
    # 전역 변수 A, B, C에 각각 3, 3, 4를 더합니다.
    global A, B, C
    A += 3
    B += 3
    C += 4

# 옥수수를 수확하는 함수를 정의합니다.
def 옥수수():
    # 전역 변수 유물에 1을 더합니다.
    global 유물
    유물 += 1

# 각각의 수확 함수를 호출하여 수확량을 누적합니다.
감자()
양파()
옥수수()

# 총 수확량을 계산합니다.
합계 = A + B + C + 유물

# 총 수확량을 출력합니다.
print('오늘 수확한 수확물:', 합계)

#출력
#오늘 수확한 수확물 : 12
````

````python
# print_text 함수 정의
def print_text():
		# text 변수에 문자열 할당
    text = "콩콩팥팥 수확의 기쁨"

		# 중첩 함수 txt 정의
    def txt():
		    # 바깥쪽 함수인 print_text의 변수인 text를 출력
        print(text)

		# 중첩 함수 호출
    txt()

# print_text 함수 호출
print_text()

#출력
# 콩콩팥팥 수확의 기쁨
````

````python
# 수확철 함수 정의
def 수확철():
    # 야채 딕셔너리 초기화
    야채 = {'A등급': 0, 'B등급': 0, 'C등급': 0}

    # 양파 함수 정의
    def 양파():
        # A등급 야채 개수 증가
        야채['A등급'] += 1

    # 대파 함수 정의
    def 대파():
        # A등급, B등급, C등급 야채 개수 증가
        야채['A등급'] += 3
        야채['B등급'] += 3
        야채['C등급'] += 4

    # 쪽파 함수 정의
    def 쪽파():
        # 금 등급 야채 추가
        야채['금'] = 1

    # 양파, 대파, 쪽파 함수 호출
    양파()
    대파()
    쪽파()

    # 야채 딕셔너리 반환
    return 야채

# 수확철 함수 호출 및 결과 출력
print(수확철())
print(sum(수확철().values()))

#출력
#{'A등급': 4, 'B등급': 3, 'C등급': 4, '금': 1}
#12
````

1. 함수는 무엇을 하기 위해 사용되나요?
    - 정답
        
        함수는 **자주 반복되는 코드를 미리 작성하여 코드의 재사용성을 높이고, 프로그램의 효율성을 증대**시키기 위해 사용됩니다.
        
    
2. 함수의 선언 방법에 대해서 설명해주세요.
    - 정답
        
        함수의 선언은 **`def`** 키워드를 사용하여 이루어집니다. 기본적인 형식은 다음과 같습니다:
        
        ```python
        def 함수이름(매개변수):
            실행할 코드
        ```
        
        **매개변수는 함수에 전달되는 입력값**을 의미하며, 필요에 따라 없어도 됩니다. **함수 내부에는 실행할 코드 블록이 위치**하게 됩니다.
        
    
3. 함수의 반환값을 어떻게 받을 수 있나요?
    - 정답
        
        함수의 반환값은 **`return`** 키워드를 사용하여 지정합니다. 반환값은 함수가 실행된 후 결과를 호출한 곳으로 반환하는데 사용됩니다.
        
        ```python
        def add(x, y):
            return x + y
        result = add(3, 5)
        print(result)
        
        #출력
        8
        ```
        
    
4. 전역 변수와 지역 변수에 차이에 대해 설명해주세요.(심화)
    - 정답
        
        전역 변수와 지역 변수의 차이는 변수의 범위에 있습니다.
        
        - **전역 변수(Global Variable)**: **프로그램 전체에서 접근 가능한 변수**로, **함수 외부에서 정의**되며 프로그램 어디서든 참조할 수 있습니다.
        - **지역 변수(Local Variable)**: **함수 내에서 정의되어 함수 내에서만 접근 가능한 변수**로, **함수가 호출될 때 생성되고 함수가 종료되면 소멸합니다. 함수 내에서만 유효**합니다.
        
        전역 변수와 지역 변수의 주요 차이는 변수의 유효 범위에 있습니다. 함수 내에서 정의된 지역 변수는 함수 내에서만 유효하며 함수 외부에서는 접근할 수 없습니다. 하지만 전역 변수는 프로그램 전체에서 접근할 수 있습니다.
