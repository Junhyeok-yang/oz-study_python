## 1. 파일 입출력의 모드들

- 읽기 모드 (r)
    - **`r`** 모드는 파일을 **읽기 전용**으로 엽니다.
    - **파일이 존재하지 않으면 오류**가 발생합니다.
- 쓰기 모드 (w)
    - **`w`** 모드는 파일을 **쓰기 전용**으로 엽니다.
    - 파일이 **존재하지 않으면 새로운 파일**을 생성합니다.
    - 이미 **파일이 존재할 경우, 기존 내용을 삭제하고 새로운 내용**을 씁니다.
- 추가 모드 (a)
    - **`a`** 모드는 파일을 **쓰기 전용**으로 엽니다.
    - 파일이 이미 **존재하면 기존 내용 끝에 추가**합니다.
    - 파일이 **존재하지 않으면 새로운 파일을 생성**합니다.

##2. 파일 만들어보기

```python
# 'test.txt' 파일을 쓰기 모드('w')로 엽니다.
# 만약 파일이 존재하지 않으면 새로 생성하고, 이미 존재한다면 내용을 덮어씁니다.
f = open('test.txt', 'w')

# 파일에 문자열 '수강생 여러분'을 씁니다.
f.write('수강생 여러분')

# 파일을 닫습니다.
f.close()
```

## 3. 파일 수정하기

```python
# 'test.txt' 파일을 쓰기 모드('w')로 엽니다.
# 만약 파일이 존재하지 않으면 새로 생성하고, 이미 존재한다면 내용을 덮어씁니다.
f = open('test.txt', 'w')

# 파일에 문자열 '개발자 여러분'을 씁니다.
f.write('수강생 여러분')

# 파일을 닫습니다.
f.close()
```

```python
# ''' '''을 이용해 여러줄을 한번에 입력할 수 있습니다.

# 'test.txt' 파일을 쓰기 모드('w')로 엽니다.
# 만약 파일이 존재하지 않으면 새로 생성하고, 이미 존재한다면 내용을 덮어씁니다.
f = open('test.txt', 'w')

# 파일에 여러 줄의 문자열을 씁니다. 여러 줄을 작성하기 위해 ''' 또는 """로 문자열을 감싸줍니다.
f.write('''
개발자 여러분
개발자 여러분
개발자 여러분
개발자 여러분
개발자 여러분
''')

# 파일을 닫습니다.
f.close()
```

## 3. for문을 이용한 파일 작성

```python
# 'test.txt' 파일을 쓰기 모드('w')로 엽니다.
# 만약 파일이 존재하지 않으면 새로 생성하고, 이미 존재한다면 내용을 덮어씁니다.
f = open('test.txt', 'w')

# 빈 문자열을 초기화합니다.
메시지 = ''

# 0부터 9까지의 범위를 반복하면서 문자열을 메시지에 추가합니다.
for i in range(10):
    메시지 += '개발자 여러분\n'

# 파일에 메시지를 씁니다.
f.write(메시지)

# 파일을 닫습니다.
f.close()
```

## 4. 파일 불러오기

```python
# 'test.txt' 파일을 읽기 모드('r')로 엽니다.
f = open('test.txt', 'r')

# 파일의 내용을 문자열 s에 저장합니다.
s = f.read()

# s를 출력합니다.
print(s)

# 파일을 닫습니다.
f.close()
```

```python
#한줄씩 읽어오기

# 'test.txt' 파일을 읽기 모드('r')로 엽니다.
f = open('test.txt', 'r')

# 파일에서 한 줄을 읽어서 line에 저장합니다.
line = f.readline()

# 읽은 한 줄을 출력합니다.
print(line)

# line의 데이터 타입을 출력합니다.
print(type(line))

# 파일을 닫습니다.
f.close()
```

## 5. while문을 이용해 파일 읽기

```python
# 'test.txt' 파일을 읽기 모드('r')로 엽니다.
f = open('test.txt', 'r')

while line:
		# 파일에서 한 줄을 읽어서 line에 저장합니다.
		line = f.readLine()

		# 저장된 line을 출력합니다.
		print(line)

# 파일을 닫습니다.
f.close()
```

## 6. 파일 내용 추가하기

```python
# 'test.txt' 파일을 추가 모드('a')로 엽니다.
f = open('test.txt', 'a')

# 파일에 '힘내자'를 추가합니다.
f.write('힘내자')

# 파일을 닫습니다.
f.close()
```

## 7. 문법의 단순화

```python
# 'test.txt' 파일을 추가 모드('a')로 열고, 파일 객체를 f로 지정합니다.
with open('test.txt', 'a') as f:
    # 파일에 '힘내자!!!'를 추가합니다.
    f.write('힘내자!!!')

#f.close()를 작성할 필요가 없다.
```

## 8. 배열을 추가하는 방법

```python
#list형태의 데이터는 문자로 형변환을 한 후 추가할 수 있다.

# 리스트 list를 정의합니다.
list = [10, 20, 30]

# 'test.txt' 파일을 추가 모드('a')로 열고, 파일 객체를 f로 지정합니다.
with open('test.txt', 'a') as f:
    # 리스트를 문자열로 변환하여 파일에 추가합니다.
    f.write(str(list))
```

## 9. JavaScript 언어 변형시키기

```python
# 리스트 list를 정의합니다.
list = [10, 20, 30]

# 'test.txt' 파일을 쓰기 모드('w')로 열고, 파일 객체를 f로 지정합니다.
with open('test.txt', 'w') as f:
    # 리스트를 문자열로 변환하여 파일에 쓰고, JavaScript 변수를 할당하는 형식으로 저장합니다.
    f.write('let test = ' + str(list))
```

## 10. list 형태 간편하게 사용하기

```python
list = ['10', '20', '30']

# 'test.txt' 파일을 쓰기 모드로 엽니다.
with open('test.txt', 'w') as f:
    # 리스트의 각 요소를 파일에 연속해서 씁니다.
    f.writelines(list)

#입력
102030
```

```python
#한줄씩 띄워서 파일에 입력하기

list = ['10', '20', '30']

# 'test.txt' 파일을 쓰기 모드로 엽니다.
with open('test.txt', 'w') as f:
    # 리스트의 각 요소를 파일에 한 줄씩 씁니다.
    for item in list:
        f.write(item + '\n')
```

## 11. csv파일 사용하기

```python
list = ['10', '20', '30']

# 'test3.csv' 파일을 쓰기 모드로 엽니다.
with open('test3.csv', 'w') as f:
    # 리스트의 각 요소를 쉼표로 구분하여 파일에 씁니다.
    f.write(','.join(map(str, list)))
```
