##시퀀스 자료형
* 문자열 :
        * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
        * 슬라이싱 ( [:] [::])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)
* list :
        * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
        * 슬라이싱 ( [:] [::])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)
        * list 지원하는 다양한 메소드(append(), insert(), del, sort(), reverse())
* tuple :
        * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제할 수 없다.
        * 값을 추가, 삭제, 변경할 수 없기 때문에, 사용하지 못하는 메소드가 많다.
        * tuple의 생성 방법은 tuple = 값, 값 또는 tuple = (값, 값)
        * tuple은 소괄호

```python
# 리스트 []대괄호, 튜플은 ()소괄호
튜플 = (1, '이', 3.0)
print(type(튜플))
print(튜플)

리스트 = [1, '이', 3.0]
print(type(리스트))
print(리스트)

튜플1 = 1, '이', 3.0
print(type(튜플1))
print(튜플1)

#하나의 값만 들어있는 튜플만 만들고 싶다면?
튜플_솔로 = (1, )
print(type(튜플_솔로))
print(튜플_솔로)
```

```python
# 값 추가, 삭제, 변경 불가...

print(튜플)
print(dir(튜플))

print(dir(list))

튜플_리스트화 = list(튜플)
print(type(튜플_리스트화))
```

```python
#번외

수강생 = '중꺽그마'

수강생_list = list(수강생)
수강생_tuple = tuple(수강생)

print(수강생_list)
print(수강생_tuple)

수강생_list[1] = '뿌'
print(수강생_list)

#join 함수
취뽀_수강생 = "".join(수강생_list)
print(취뽀_수강생)

```

```python
# 튜플과 리스트의 공통점
print(튜플)
# 길이를 구할 수 있어요
len(튜플)
#곱하기 연산으로 요소를 복사 할 수 있다.
튜플 * 3
#튜플은 요소들끼리 더할 수 있다.
튜플 + 튜플
#튜플은 인덱싱이 가능하다
튜플[2]
#튜플의 인덱스
튜플.index(3.0)
#튜플 요소의 개수
튜플.count(1)
#튜플도 슬라이싱이 가능하다
튜플[1:3]
```

#tuple 왜 쓸까?

1.3go가 불가하기 때문(넣고, 삭제하고, 바꾸고)
        
        코딩을 하는 경우 변경되면 안되는 변수를 건드리는 경우  
        또는 값이 변경되면 안되는 중요한 데이터를 넣어서 보안적으로 관리하기 좋다.(은행계좌, 비밀번호 등등...)

2.list는 tuple보다 무겁다

        list가 단지 기능이 많아서 무거운걸까?
        파이썬은 c로 만들어졌다.

        파이썬은 동적배열을 지원한다.
        - list = [1, 2, 3, 4] 생성해서 4개의 값을 모두 넣었다.
        - 문제는 그 다음 값을 추가할때 이미 공간이 다 찼기 때문에 추가 할당을 진행해야한다.
          추가 할당 시 메모리 크기는 4, 8, 16, 25...더블링을 해준다.

3.리스트는 가변, 튜플은 불변이다.

        값이 고정되어야 하거나 크기가 작을 경우 튜플을 권장한다.
