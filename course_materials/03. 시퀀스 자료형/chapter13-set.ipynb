##시퀀스 자료형
* 문자열 :
        * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
        * 슬라이싱 ( [:] [::])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)
* list :
        * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
        * 슬라이싱 ( [:] [::])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)
        * list 지원하는 다양한 메소드(append(), insert(), del, sort(), reverse())
* tuple :
        * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제할 수 없다.
        * 값을 추가, 삭제, 변경할 수 없기 때문에, 사용하지 못하는 메소드가 많다.
        * tuple의 생성 방법은 tuple = 값, 값 또는 tuple = (값, 값)
        * tuple은 소괄호
* dict(딕셔너리)
        * list, tuple, str, range(범위 안에 연속된 데이터를 만드는 시퀀스)
          -공통적인 특징 : 데이터의 연관성이 없다.
        * dict는 여러개의 값이 일렬로 정렬되면서도 값끼리의 연관성이 존재
        (예시 : 게임 캐릭터 힘:30 지능:20 체력:40 민첩:90)
        * 생성 방법은 중괄호
* set
        * set은 수학의 집합을 의미
        * 순서가 없다.
        * 값의 중복을 허락하지 않는다.
        * set 생성 방법도 중괄호


```python
채소 = {'당근', '배추', '대파', '양파'}
type(채소)
print(채소)

#순서가 없고, 출력하면 요소가 다르게 나오는 것도 특징.

# 세트를 만드는 다양한 방법
세트 = set()
print(type(세트))

세트2 = {}
print(type(세트2))
```

```python
# 중복을 제일 싫어하는 set

사과_리스트 = list('apple')
사과_튜플 = tuple('apple')
사과_세트 = set('apple')
print(사과_리스트)
print(사과_튜플)
print(사과_세트)
```

```python
# set()자료형의 데이터 추가, 삭제, 변경

채소

#set () 함수의 데이터 삽입
채소.add('토마토')

#update로 하면 요소를 분리해서 입력
채소.update('깻잎')
채소

#set() 함수의 데이터 삭제
채소.remove('잎')
채소

채소.discard("깻")
채소

#remove는 없는 값을 제거하면 에러가 생기지만, discard는 에러가 생기지 않음

#pop은 임의로 삭제하고 반환해줌
채소.pop()
채소
반환채소 = 채소.pop()
print(반환채소)
```

```python
# set = 집합, 합집합, 교집합, 차집합 등등...
# 대표적으로 합집합, 교집합, 차집합

채소1 = {'당근', '양파', '오이', '배추'}
채소2 = {'양파', '오이', '대파', '가지'}

#합집합(union)
채소1 | 채소2
print(set.union(채소1, 채소2))

#교집합(intersection)
채소1 & 채소2
print(set.intersection(채소1, 채소2))

#차집합(difference
채소1 - 채소2
print(set.difference(채소1, 채소2))

#대칭차집합
채소1 ^ 채소2
print(set.symmetric_difference(채소1, 채소2))
```
