##시퀀스 자료형
* 문자열 :
        * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
        * 슬라이싱 ( [:] [::])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)
* list :
        * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
        * 슬라이싱 ( [:] [::])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)
        * list 지원하는 다양한 메소드(append(), insert(), del, sort(), reverse())
* tuple :
        * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제할 수 없다.
        * 값을 추가, 삭제, 변경할 수 없기 때문에, 사용하지 못하는 메소드가 많다.
        * tuple의 생성 방법은 tuple = 값, 값 또는 tuple = (값, 값)
        * tuple은 소괄호
* dict(딕셔너리)
        * list, tuple, str, range(범위 안에 연속된 데이터를 만드는 시퀀스)
          -공통적인 특징 : 데이터의 연관성이 없다.
        * dict는 여러개의 값이 일렬로 정렬되면서도 값끼리의 연관성이 존재
        (예시 : 게임 캐릭터 힘:30 지능:20 체력:40 민첩:90)
        * 생성 방법은 중괄호


```python
#list [], tuple (), dict {}

충무로 = {'힘':30, '지능':20, '체력':40, '민첩':90}
충무로

#{키 : 값}
#{key1:value1 key2:value2}

# 시퀀스 데이터는 인덱싱을 이용해 데이터를 찾는다.

#충무로[1]
충무로['힘']
print(f'충무로님은 힘이 {충무로["힘"]}입니다.')
```

```python
#딕셔너리를 만드는 다양한 방법

#빈 딕셔너리 만들기
딕셔너리 = {}
print(딕셔너리)

딕셔너리2 = dict()
print(딕셔너리2)

#값을 넣어 만드는 방법
#dict()로 생성시 값 넌는 기호 "", : 주의하기
딕셔너리3 = dict(힘=30, 지능=20, 체력=40, 민첩=90)
print(딕셔너리3)

#zip() : 병렬처리를 도움
딕셔너리4 = dict(zip(['힘', '지능', '체력', '민첩'], [30, 20, 40, 90]))
print(딕셔너리4 )
```

```python
# 값의 중복은 허락되는데, 키의 중복은 허락하지 않습니다. 중복되면 맨 뒤의 값으로 나옴.

충무로 = {'힘':30, '지능':20, '체력':40, '민첩':90, '힘': 40, '지구력':90}
충무로

'체력' in 충무로
```

```python
#딕셔너리에 다양한 메소드

print(충무로.keys())
print(충무로.values())

print(충무로.items())
print(type(충무로.items()))

리스트화된_충무로 = list(충무로.items())
리스트화된_충무로[0]
#리스트화 시키고 인덱싱도 가능
```

```python

#리펙토링하기

#레거시 : 누군가 남겨두고간 코드
#리펙토링 : 결과의 변경 없이 코드의 구조를 재조정하는 것.
#

#붕어빵 이름
#fishcake__bean = "붕어빵(팥)"
#fishcake__cream = "붕어빵(슈크림)"
#fishcake__noodle = "붕어빵(잡채)"
fishcake__bean, fishcake__cream, fishcake__noodle = input('붕어빵(팥), 붕어빵(슈크림), 붕어빵(잡채)').split(",")

#붕어빵 개수
#bean__count = 0
#cream__count = 2
#noodle__count = 5
bean__count, cream__count, noodle__count = map(int, input().split(','))

#붕어빵 총 개수
total__count = bean__count + cream__count + noodle__count

#붕어빵 가격 출력
bean__price = 2000 * bean__count
cream__price = 2500 * cream__count
noodle__price = 3000 * noodle__count
#붕어빵 총 가격
total__price = bean__price + cream__price + noodle__price


print(f'주문이 완료되었습니다. {fishcake__bean} {bean__count}개, {fishcake__cream} {cream__count}개, {fishcake__noodle} {noodle__count}개 총 {total__count}개 결제 금액은 {total__price}원 입니다.')
```
