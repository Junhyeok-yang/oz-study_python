###list

* list란? 원소들이 연속적으로 저장되는 형태의 자료형
* 시퀀스 자료형
  * str, list, tuple, dict, set 등
* 만드는 방법은 대괄호로 묶어주고, 값들은 콤마로 구분.

```python
# 첫 리스트

리스트 = []
print(리스트)

리스트 = [23, "글자", 3.14, True, False]
print(리스트)
```

```python
#리스트 다시 만들어보기.
리스트 = ['이', '렇', '게', '만', '들', '있','음']
print(리스트)
#빈 리스트 만들기
리스트2 = list()
print(리스트2)

#range? 연속된 숫자를 생산하는 기능
range(10)
```

```python
#list(rang(시작,끝))
리스트_매직_들어간다 = list(range(0,20))
print(리스트_매직_들어간다)


#list(range(시작, 끝, 증가))
리스트_매직2 = list(range(0, 20, 2))
print(리스트_매직2)

리스트_매직3 = list(range(20, 0, -1))
print(리스트_매직3)

리스트_매직4 = list(range(20, 0))
print(리스트_매직4)
```

```python
# 끝나지 않는 리스트 매직
# 리스트를 이용해 변수에 값 넣기

a, b, c = ['들', '어', '가']
print(a, b, c)

d, e, f = [1, 3, 5]
print(d, e, f)

지갑 = [10000, 5000, 1000]
만원, 오천원, 천원 = 지갑
print(f'{만원} 2장, {오천원} 3장, {천원} 1장')
```

```python
#슬라이싱

로또 = [3, 5, 15, 33, 41, 44]
#로또 변수에 있는 리스트의 인덱스 1번부터 2번까지의 값을 출력해보기
print(로또[0:2])

#로또 변수의 마지막 위치 출력해보기v
print(로또[-1])


#반대로 출력해보기
print(로또[::-1])

#로또 리스트 안에 33이 있는지 확인해보기
print(로또.count(33))
print(33 in 로또)

#로또 리스트를 로또2 리스트와 합쳐보기
로또2 = [2, 12, 15, 24, 33, 39]
로또3 = 로또 + 로또2
print(로또3)
print(로또 + 로또2)

#range를 이용해 1부터 10 사이에 짝수만 들어있는 짝수(변수명) 리스트 만들기
짝수 = list(range(2,11,2))
print(짝수)

#짝수 리스트에 들어있는 값을 2배 늘리고 다시 짝수 변수에 담기V
짝수 = 짝수 * 2
print(짝수)

#짝수 리스트에 들어있는 요소의 개수 구하기
print(len(짝수))

#짝수 리스트의 3번째 인덱스 출력하기V
print(짝수[3])

#len()함수를 이용해 인덱스 마지막 값을 출력하기V
print(짝수[len(짝수)-1])

```

```python
# list 값 넣고, 바꾸고, 지우고

mzfood = ['숙주', '분모자', '마라', '소세지', '소고기', '옥수수면']

# 값을 넣는 방법 2가지 :
#append() : 마지막 요소에 값을 넣음.
#insert() : 원하는 곳에 값을 넣음

mzfood.append("고수")
print(mzfood)

mzfood.insert(0, "탕후루")
print(mzfood)

#값 지우고
del mzfood[-1]
print(mzfood)

del mzfood[0]
print(mzfood)

# 값 바꾸고
mzfood[-2] = '양고기'
print(mzfood)

# 문자열 시퀀스여도 이용할 수 없다.
인사 = '반가워'
인사[2]='웡'
```

```python
print(mzfood)

#거꾸로
mzfood.reverse()
print(mzfood)

#오름차순
mzfood.sort()
print(mzfood)


# reverse 왜 print문을 쓰면 none이 나올까
print(mzfood.sort())
print(mzfood.reverse())
```
