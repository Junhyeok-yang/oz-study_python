```python
#형변환을 하면 input()으로 입력을 해도 형변환 되어 출력
#int(), float(), str(), bool()

정수 = input()
정수 = int(정수)
print(정수, type(정수))

실수 = input()
실수 = float(실수)
print(실수, type(실수))
```

```python
정수 = int(input())
print(정수, type(정수))

실수 = float(input())
print(실수, type(실수))
```

그 외 입력값의 형변환이 가능한 종류

 * dict()

```python
넌센스, 정답 = input().split()
print(f'넌센스 퀴즈 : {넌센스}')
print(f'정답 : {정답}')
#split() 띄어쓰기 하나만 인식
```

```python
#쉼표로 구분
넌센스, 정답 = input().split(",")
print(f'넌센스 퀴즈 : {넌센스}')
print(f'정답 : {정답}')
```

```python
#시퀀스 자료형에 다 지원되는 메소드
번호1, 번호2, 번호3, 번호4 = input().split(",")

번호1 = int(번호1)
번호2 = int(번호2)
번호3 = int(번호3)
번호4 = int(번호4)

print(type(번호1), 번호2, 번호3, 번호4)
```

```python
#map()
번호1, 번호2, 번호3, 번호4 = map(int, input().split(","))
print(type(번호1), 번호2, 번호3, 번호4)
```

```python
#레거시 : 누군가 남겨두고간 코드
#리펙토링 : 결과의 변경 없이 코드의 구조를 재조정하는 것.
#

#붕어빵 이름
#fishcake__bean = "붕어빵(팥)"
#fishcake__cream = "붕어빵(슈크림)"
#fishcake__noodle = "붕어빵(잡채)"
fishcake__bean, fishcake__cream, fishcake__noodle = input('붕어빵(팥), 붕어빵(슈크림), 붕어빵(잡채)').split(",")

#붕어빵 개수
#bean__count = 0
#cream__count = 2
#noodle__count = 5
bean__count, cream__count, noodle__count = map(int, input().split(','))

#붕어빵 총 개수
total__count = bean__count + cream__count + noodle__count

#붕어빵 가격 출력
bean__price = 2000 * bean__count
cream__price = 2500 * cream__count
noodle__price = 3000 * noodle__count
#붕어빵 총 가격
total__price = bean__price + cream__price + noodle__price


print(f'주문이 완료되었습니다. {fishcake__bean} {bean__count}개, {fishcake__cream} {cream__count}개, {fishcake__noodle} {noodle__count}개 총 {total__count}개 결제 금액은 {total__price}원 입니다.')
```
