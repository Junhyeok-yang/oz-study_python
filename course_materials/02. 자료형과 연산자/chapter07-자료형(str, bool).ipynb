### 자료형 종류

```python
#자료형 종류
#1. int(integer) : 정수
#2. float(float) : 실수
#3. str(string) : 문자
#4. bool(bool, boolean) : 참 또는 거짓, true 또는 false, 1 또는 0
---
#5. list
#6. tuple
#7. dict
#8. set
```

```python
이름 = "김오즈"
나이 = 30
키 = 180.3
소속 = "마법중"
```

```python
print(type(이름))
print(type(나이))
print(type(키))
print(type(소속))
```

```python
print(float(5))
print(int(5.0))
print("5", "5.0")
print(type("5"), type("5.0"))
```

```python
생년월일 = "2000.12.25"
#2000 12 25
년 = 생년월일[0:4]
월 = 생년월일[5:7]
일 = 생년월일[8:10]

print(f'{년} {월} {일}')
```

```python
숫자 = "1234567890"
#전체 출력
print(숫자[:])
print(숫자[::])

#거꾸로 출력
print(숫자[::-1])

#첫번째 인덱스 기준으로 2칸씩 띄어쓰기 출력하기
print(숫자[::2])

#부분 출력하기
print(숫자[4:1:-1])
print(숫자[1:4])
```

```python
라라랜드 = "사람들은 다른 사람들의 열정에 끌리게 되어있어 자신이 잊은 걸 상기시켜 주니까"

print(라라랜드.index("다"))
print(라라랜드.count("사"))
print(라라랜드[:])
print(라라랜드[::])
print(라라랜드[:-1])
print(라라랜드[::-1])
print(라라랜드[0:3])
print(라라랜드[3:0:-1])
print(라라랜드[::-3])
print(라라랜드[::-2])
print(len(라라랜드))
```
