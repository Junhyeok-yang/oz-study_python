```python
# 중요
# ==, is

print(3 is 3)
print(3 == 3)

print(3.0 is 3)
print(3.0 == 3)

print(id(3.0))
print(id(3))
```

미션 : 값 비교에는 is를 쓰면 안되는데 그 이유를 찾아서 기술 블로그에 작성하기

```python
#본격적인 논리연산자
# True : 1 / False : 0

print(True and True) # 1 * 1 = 1
print(True and False) # 1 * 0 = 0
print(False and True) # 0 * 1 = 0
print(False and False) # 0 * 0 = 0

print(True or True) # 1 + 1 = 1
print(True or False) # 1 + 0 = 1
print(False or True) # 0 + 1 = 1
print(False or False) # 0 + 0 = 0
```

```python
#굉장히 중요한 내용
#단락 평가

print(False and True)
print(False and "힘들어요")
#힘들어요라는 뒤에꺼는 보지 않음. 이것이 단락 평가

print(True and False)

print(True or "힘들어요")
print(True and "힘들어요")

#이건 기술블로그에 좀 넣어야할듯 이해 안됨.
```
