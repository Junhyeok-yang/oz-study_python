## 1. 튜플

튜플은 list와 비슷한 시퀸스 자료형입니다.

단, list와의 차이점은 요소를 바꿀 수 없다는 점입니다.

튜플을 생성할 때는 소괄호”()”를 이용합니다.

ex) 변수명 = (데이터1, 데이터2, 데이터3)

````python
oz_tuple = (1, 2, 3)  # 튜플 oz_tuple을 정의하고 초기화

print(oz_tuple)  # 전체 튜플을 출력 (출력: (1, 2, 3))
print(oz_tuple[0])  # 첫 번째 요소를 출력 (출력: 1)
print(oz_tuple[1])  # 두 번째 요소를 출력 (출력: 2)
print(oz_tuple[2])  # 세 번째 요소를 출력 (출력: 3)

# 출력
(1, 2, 3)
1
2
3
````

## 2. 튜플 데이터 변경

튜플은 **데이터 변경이 불가능**합니다.

````python
oz_tuple = (1, 2, 3)
oz_tuple = 5

# 출력
TypeError: 'tuple' object does not support item assignment
````

## 3. 튜플 데이터 할당

튜플은 괄호 없이 여러 값을 할당할 수 있습니다.

````python
oz_tuple = 1, 2, 3, 4

print("oz_tuple: ", oz_tuple)
print("oz_tuple type: ", type(oz_tuple))

# 출력
oz_tuple: (1, 2, 3, 4)
oz_tuple type: <class 'tuple'>
````

````python
x, y, z = 1, 2, 3
print(x)
print(y)
print(z)

# 출력
1
2
3
````

## 4. 튜플 값 교환

튜플 내부의 값을 변경할 수는 없지만 값을 교환할 수는 있습니다.

````python
x, y = 1, 2

print("x :", x)
print("y :", y)
print()

x,y = y,x
print("x :", x)
print("y :", y)

# 출력
x: 1
y: 2

x: 2
y: 1
````

## 5. 튜플의 값 리턴

튜플을 사용하면 여러 개의 값을 리턴할 수 있습니다.

````python
def oz_tuple():
	return (1, 2)

x, y = oz_tuple()

print("x: ", x)
print("y: ", y)

# 출력
x: 1
y: 2
````

## 6. 튜플을 이용한 내부함수

````python
# 튜플을 이용한 내부함수들
# enumerate()의 1, fruit는 (1, fruit)이나 괄호를 생략한 상태입니다.

# 튜플 (1, 2, 3, 4)의 각 요소와 해당 요소의 인덱스를 반복하여 출력
for index, fruit in enumerate((1, 2, 3, 4)):
    print(index, fruit)  # 현재 인덱스와 해당 인덱스의 요소를 출력

# 출력
0 1
1 2
2 3
3 4
````

````python
# divcode() 사용 시 튜플 형태로 몫과 나머지를 반환
a, b = 7, 3  # 변수 a와 b를 각각 7과 3으로 초기화
x, y = divmod(a, b)  # divmod 함수를 사용하여 a를 b로 나눈 몫과 나머지를 구함

print("몫: ", x)  # 계산된 몫 x를 출력
print("나머지: ", y)  # 계산된 나머지 y를 출력

# 출력
몫: 2
나머지: 1
````

1. 튜플의 특징에 대해 설명해주세요.
    - 정답
        1. **불변성 (Immutable):** 튜플은 한 번 생성되면 그 안의 요소를 변경할 수 없습니다. 즉, 요소의 값을 추가, 수정, 삭제할 수 없습니다. 이를 통해 데이터의 무결성을 유지할 수 있습니다.
            
            ```python
            oz_tuple = (1, 2, 3)
            oz_tuple[0] = 5  # TypeError: 'tuple' object does not support item assignment
            ```
            
        2. **소괄호 사용:** 튜플은 소괄호 ()를 사용하여 생성합니다.
            
            ```python
            oz_tuple = (1, 2, 3)
            ```
            
        3. **데이터 할당:** 튜플은 여러 값을 콤마로 구분하여 할당할 수 있으며, 소괄호를 생략할 수도 있습니다.
            
            ```python
            oz_tuple = 1, 2, 3, 4
            ```
            
        4. **다양한 자료형:** 튜플은 다양한 자료형의 요소를 가질 수 있습니다.
            
            ```python
            oz_tuple = (1, "hello", 3.14)
            ```
            
        5. **데이터 교환:** 튜플을 사용하면 값의 교환이 간단해집니다.
            
            ```python
            oz_tuple = (1, "hello", 3.14)
            ```
            
        6. **여러 값 리턴:** 함수에서 여러 개의 값을 리턴할 때 튜플을 사용하면 편리합니다.
            
            ```python
            def oz_tuple():
                return (1, 2)
            
            x, y = oz_tuple()
            ```
