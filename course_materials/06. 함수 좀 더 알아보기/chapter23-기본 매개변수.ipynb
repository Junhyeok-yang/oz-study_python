## 1. 기본 매개변수란?

Child 클래스 내부에는 아무것도 없는데, Parent 클래스의 상속을 받았으므로

함수를 호출할 때 해당 매개변수에 **인수를 전달하지 않으면 기본 값이 매개변수에 사용**됩니다.

기본 매개변수를 사용하면 함수 호출시에 모든 매개변수에 값을 전달하지 않아도 됩니다.

## 2. 기본 매개변수의 활용

```python
def oz(value, n=2):
    # value: 출력할 문자열을 나타내는 일반 매개변수입니다.
    # n: 출력할 횟수를 나타내는 기본 매개변수입니다. 만약 명시되지 않으면 기본값으로 2가 사용됩니다.
    for i in range(n):
        # value를 n번 출력합니다.
        print(value)

# oz 함수를 호출하고 "일반 매개변수 다음에 기본 매개변수를 배치해야 합니다." 문자열을 전달합니다.
# n이 명시되지 않았으므로 기본값으로 2가 사용됩니다.
oz("일반 매개변수 다음에 기본 매개변수를 배치해야 합니다.")

#출력
일반 매개변수 다음에 기본 매개변수를 배치해야 합니다.
일반 매개변수 다음에 기본 매개변수를 배치해야 합니다.
```

기본 매개변수의 값을 지정해놨더라도, 함수를 **호출할 때 매개변수의 값을 변경**할 수 있습니다.

```python
def oz(value, n=2):
    # value: 출력할 문자열을 나타내는 일반 매개변수입니다.
    # n: 출력할 횟수를 나타내는 기본 매개변수입니다. 명시되지 않을 경우 기본값은 2입니다.
    for i in range(n):
        # value를 n번 출력합니다.
        print(value)

# oz 함수를 호출하고 "일반 매개변수 다음에 기본 매개변수를 배치해야 합니다." 문자열을 전달합니다.
# n 매개변수에 10을 전달하여 해당 문자열을 10번 출력합니다.
oz("일반 매개변수 다음에 기본 매개변수를 배치해야 합니다.", n=10)

#출력
일반 매개변수 다음에 기본 매개변수를 배치해야 합니다. * 10
```

**기본 매개변수 뒤에는 일반 매개변수가 올 수 없습니다.**

```python
def oz(n=2, value):
    # 함수 oz 정의 시, 기본 매개변수를 일반 매개변수보다 뒤에 위치시켜야 합니다.
    for i in range(n):
        print(value)

oz("일반 매개변수 다음에 기본 매개변수를 배치해야 합니다.")


#출력
SyntaxError: non-default argument follows default argument
```

1. 기본 매개 변수의 역할을 무엇인가요?
    - 정답
        
        기본 매개변수의 역할은 함수를 호출할 때 해당 매개변수에 인자가 전달되지 않을 경우에 사용됩니다.
        
        즉, 함수를 호출할 때 인자를 전달하지 않으면 기본값이 매개변수에 할당되어 함수가 실행됩니다.
        
    
2. 기본 매개 변수를 사용하기 위한 문법은 어떻게 되나요?
    - 정답
        
        기본 매개변수를 사용하기 위한 문법은 매개변수 선언 시에 해당 매개변수에 기본값을 할당하는 것입니다. 기본값은 등호(**`=`**)를 사용하여 지정하며, 함수의 매개변수 리스트에서 일반 매개변수 뒤에 나오게 됩니다.
        

1. 기본 매개 변수의 사용 시 제약 조건에 대해 설명해주세요.
    - 정답
        
        매개변수는 반드시 기본값이 없는 매개변수 뒤에 위치해야 합니다. 즉, 함수 정의 시에는 일반 매개변수가 먼저 와야 하고, 그 뒤에 기본 매개변수가 올 수 있습니다. 이를 어기게 되면 SyntaxError가 발생합니다.
