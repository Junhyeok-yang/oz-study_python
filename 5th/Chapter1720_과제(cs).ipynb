[문제1]

객체지향에 대해서 설명해주세요

**객체지향 프로그래밍(OOP, Object-Oriented Programming)**은 **데이터와 그 데이터를 처리하는 기능(메서드)**을 하나로 묶어 **객체(object)**라는 단위로 관리하는 프로그래밍 방식. 이 방식은 코드를 더 구조화하고, 재사용 가능하게 만들기 위해 만들어짐.

## 1. 객체지향 프로그래밍의 기본 개념  

### 1.1. 객체(Object)  
객체는 **데이터(속성)**와 **동작(메서드)**를 묶어 놓은 하나의 단위.  
예를 들어, 자동차 객체를 생각해 보면, 속성으로는 색깔, 모델이 있고, 동작(메서드)으로는 달리다, 멈추다 같은 기능.  

### 1.2. 클래스(Class)
클래스는 객체를 만들기 위한 설계도.  
클래스를 기반으로 여러 개의 **객체(인스턴스)**를 만들 수 있음.
예를 들어, 자동차라는 클래스를 정의하면, 그 클래스로부터 다양한 자동차 객체를 만들 수 있음. 각각의 객체는 다른 속성을 가질 수 있지만, 같은 동작을 공유할 수 있음.  

### 1.3. 인스턴스(Instance)
인스턴스는 클래스에서 만들어진 실제 객체.  
하나의 클래스에서 여러 개의 인스턴스를 만들 수 있음.
예를 들어, Car라는 클래스에서 만든 각각의 자동차 객체는 서로 다른 인스턴스.  

## 2. 객체지향 프로그래밍의 4대 특징  

### 2.1. 캡슐화(Encapsulation)  
캡슐화는 데이터를 외부에서 직접 접근하지 못하도록 숨기고, 메서드를 통해서만 접근하도록 만드는 개념.  
객체의 내부 상태는 메서드를 통해서만 변경할 수 있도록 제한하는데, 이렇게 하면 데이터가 의도하지 않게 변경되는 것을 방지할 수 있음.

### 2.2. 상속(Inheritance)  
상속은 기존 클래스의 속성과 메서드를 물려받아 새로운 클래스를 만드는 것.  
예를 들어, 동물(Animal)이라는 기본 클래스가 있으면, 이를 상속받아 개(Dog)라는 파생 클래스를 만들 수 있음. 개 클래스는 동물 클래스의 기능을 모두 물려받고, 추가로 자신만의 기능도 가질 수 있음.

### 2.3. 다형성(Polymorphism)  
다형성은 같은 이름의 메서드가 여러 클래스에서 다른 방식으로 동작할 수 있는 것.  
예를 들어, Animal 클래스는 sound()라는 메서드를 가지고 있고, 이를 상속받은 Dog 클래스와 Cat 클래스는 sound() 메서드를 각각 다르게 구현할 수 있음. Dog는 "멍멍", Cat은 "야옹".

### 2.4. 추상화(Abstraction)  
추상화는 불필요한 정보는 숨기고, 필요한 부분만 보여주는 것.  
예를 들어, 자동차 클래스에서 우리가 달리다, 멈추다 같은 동작은 필요하지만, 내부에서 엔진이 어떻게 돌아가는지까지는 알 필요가 없음. 중요한 동작만 노출하는 것이 추상화.  

## 3. 객체지향 프로그래밍의 장점  
* 재사용성: 클래스를 한 번 정의하면, 여러 인스턴스를 만들어 재사용할 수 있음.  
* 유지보수 용이: 코드가 객체 단위로 나누어져 있어서 수정하기가 쉬움.  
* 확장성: 새로운 기능을 추가할 때 기존 클래스를 상속받아 쉽게 확장할 수 있음.  

## 4. 객체지향 프로그래밍의 단점
* 복잡성: 객체 지향은 구조가 복잡할 수 있어서, 작은 프로젝트에서는 과도할 수 있음.
* 속도 문제: 절차 지향 프로그래밍보다 처리 속도가 느릴 수 있어, 특히 간단한 작업에서는 불리할 수 있음.  

## 요약:  
**객체지향 프로그래밍(OOP)**은 클래스와 객체를 기반으로 프로그램을 설계하고 구현하는 방법. 캡슐화, 상속, 다형성, 추상화 같은 특성을 통해 유지보수성, 확장성, 재사용성을 높이는 것이 목적.

init 메소드에 대해 설명해주세요

* __init__ 메서드는 클래스에서 인스턴스가 생성될 때 호출되는 생성자 메서드로, 인스턴스를 초기화하는 데 사용.
* __init__을 사용하면 인스턴스 속성을 설정하고, 필요한 초기 작업을 처리할 수 있음.
* 클래스가 인스턴스화될 때 자동으로 호출되므로, 명시적으로 호출할 필요는 없음.

[문제3]

클래스 상속에 대해 설명해주세요

클래스 상속은 **기존 클래스(부모 클래스)**의 속성과 메서드를 **새로운 클래스(자식 클래스)**가 물려받아 사용하는 개념. 상속을 통해 기존 코드를 재사용하고, 자식 클래스에서 부모 클래스의 기능을 확장하거나 수정할 수 있음. 파이썬에서 상속은 코드의 중복을 줄이고, 유지보수를 쉽게 하기 위해 사용됨.

[문제4]

클래스 다중 상속과 관련된 MRO에 대해 설명해주세요

**MRO(Method Resolution Order)**는 다중 상속에서 메서드나 속성을 찾을 때 클래스가 상속을 탐색하는 순서.  
파이썬에서 클래스가 다중 상속을 받을 때, 어떤 순서로 부모 클래스들을 탐색할지 결정하는 규칙이 필요하고, 그 규칙이 바로 MRO.

[문제5]

Self는 어떤 키워드 인가요?

**self**는 인스턴스 자신을 가리키는 매개변수로, 클래스 내 메서드가 해당 인스턴스에 접근하고 동작할 수 있도록 도와줌.

[문제6]

ZeroDivisionError,  IndexError 외에 3가지 이상의 파이썬 에러에 대해 설명해주세요


**TypeError**: 데이터 타입이 부적절한 경우 발생합. 예를 들어, 서로 다른 데이터 타입 간에 연산을 시도하는 경우 발생.  
**ValueError**: 데이터의 값이 부적절한 경우 발생. 예를 들어, 잘못된 형식의 값을 변환하려고 할 때 발생.  
**NameError**: 정의되지 않은 변수를 참조하려고 할 때 발생.  
**KeyError**: 사전에서 존재하지 않는 키를 사용하려고 할 때 발생.  
**FileNotFoundError**: 파일을 찾을 수 없는 경우 발생.
