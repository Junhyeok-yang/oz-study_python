[문제1]

Python의 변수 할당 개념을 다른 언어와 비교해 설명해주세요


파이썬의 변수를 이름표, C나 자바의 변수를 박스라고 생각하면 된다.

예를 들어 x = 10 이라는 변수가 있다고 가정해보자.

파이썬에서 x는 이름표와 같아서 10에 붙였던 이름표를 20에 붙이면 x는 20을 가리킨다. 이 때 10은 여전히 남아있어서 값 자체는 변하지 않고 이름표만 바꾼 것이라 생각하면 된다.

반면, 자바나 C언어에서는 변수가 박스와 같아서 박스에 10이라는 값을 넣었다가 이를 20으로 바꾸기 위해서는 10이라는 값을 버리고 20이라는 값을 넣어야 한다.

즉, 파이썬에서의 변수는 손가락으로 뭔가를 가리키 듯이 값을 가리키는 것이고, 자바와 C에서는 손으로 물을 푸듯이 값을 저장하는 것이다.

따라서, 딱 맞는 숫자를 찾지 못하고 비슷한 값으로 계산하여 오차가 발생하는 것이다.
