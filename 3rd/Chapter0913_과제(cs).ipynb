[문제1]

파이썬 2.7 버전과 파이썬 3에서 range 차이를 설명해주세요

## 파이썬 2.7과 파이썬 3에서 range()함수는 동작 방식과 메모리 사용 측면에서 차이를 보여준다.


### 파이썬 2.7에서 range()함수는 리스트를 반환한다. 


예시
>print(range(3))

>[0, 1, 2]

여기서 range(3)은 0부터 2까지 리스트를 한 번에 만들어서 반환한다. 즉, 리스트가 크면 클수록 메모리 사용량도 커지기 때문에 대량의 데이터에서는 비효율적이다.


### 파이썬 3에서 range()함수는 range객체를 반환한다.


예시
>print(range(3))

>range(0, 5)

여기서 반환된 객체는 지연 평가 방식을 사용하기 때문에, 숫자를 미리 생성하지 않고, 반복문에서 필요한 순간에만 값을 생성한다. 
이러한 방식 덕분에 파이썬 3의 range()는 메모리 사용량이 훨씬 적어서 큰 숫자 범위를 다루더라도 메모리에 큰 부담을 주지 않는다.
파이썬 2.7에서는 이러한 지연 평가 방식의 생성을 xrange()가 수행한다.


## 요약


* 파이썬 2.7에서 range()는 리스트를 반환하고, 메모리를 많이 사용한다.
* 파이썬 3에서는 range()가 range 객체를 반환하며, 필요한 값만 즉시 생성해 효율적으로 메모리를 사용한다.
* 파이썬 3에서는 xrange()가 사라지고, range()가 xrange()의 역할을 한다.
* 파이썬 3에서 파이썬 2.7처럼 리스트로 반환받고 싶다면 list(range())로 받으면 된다.

[문제2]

Python의 얕은 복사와 깊은 복사에 대해서 설명하세요

## 얕은 복사와 깊은 복사


### 얕은 복사 (Shallow Copy)
얕은 복사는 객체의 최상위 레벨의 데이터만 복사하고, 그 안에 포함된 객체는 참조만 하는 복사 방식이다.
즉, 하위 객체를 원본과 공유하게 된다. 이 때문에 복사본을 수정하면 원본 객체도 수정된다.  

예시  


> 최상위 = [['하', '위'], ['객', '체']]
> 얕은_복사 = copy.copy(최상위)

> 얕은_복사[0][0] = '바'

>print("원본 :", 최상위)
>print("복사본 :", 얕은_복사)

>원본 : [['바', '위'],['객', '체']]
>복사본 : [['바', '위'],['객', '체']]

  
* copy.copy()는 얕은 복사를 수행한다.
* 리스트의 최상위는 새롭게 복사되지만, 리스트 내부의 요소들은 원본과 공유하기 때문에 복사본을 수정하면 원본 리스트도 변경된다.
  
  
### 깊은 복사 (Deep Copy)
깊은 복사는 객체와 그 안에 포함된 모든 객체를 재귀적으로 완전히 복사하는 방식이다.
즉, 복사된 객체는 원본과 독립적인 새로운 객체가 된다. 따라서, 복사본을 수정해도 원본 객체는 수정되지 않는다.  
  
예시
  

> 최상위 = [['하', '위'], ['객', '체']]
> 깊은_복사 = copy.deepcopy(최상위)

> 깊은_복사[0][0] = '바'

>print("원본 :", 최상위)
>print("복사본 :", 깊은_복사)

>원본 : [['하', '위'],['객', '체']]
>복사본 : [['바', '위'],['객', '체']]
  

* copy.deepcopy()는 깊은 복사를 수행한다.
* 리스트와 그 안에 포함된 모든 하위 리스트까지 완전히 새로운 객체로 복사됐기 때문에, 복사본을 수정해도 원본 리스트는 변경되지 않는다.
  
  
### 언제 사용하나?

* 얕은 복사 : 객체의 하위 요소들이 바뀔 가능성이 없거나, 하위 요소들을 원본과 공유하고 싶은 경우,
* 깊은 복사 : 객체 내부의 모든 요소가 독립적이어야 하거나, 복사본이 원본에 영향을 주지 않도록 하고 싶은 경우.
